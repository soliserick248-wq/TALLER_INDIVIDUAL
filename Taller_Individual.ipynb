```python
"""
Darwin Rodríguez Merchán
===========================================================
LABORATORIO GUIADO
Modelamiento y Simulación
Generación de Números Pseudoaleatorios
y Variables Pseudoaleatorias
===========================================================

INSTRUCCIONES GENERALES
-----------------------------------------------------------
1. Lea cuidadosamente cada ejercicio.
2. Complete únicamente las secciones marcadas con TODO.
3. Ejecute el archivo después de cada ejercicio.
4. Analice los resultados obtenidos.
5. Responda las preguntas planteadas por el docente.
6. NO ELIMINE COMENTARIOS NI ESTRUCTURAS BASE.
7. Puede investigar documentación oficial o teoría adicional.

OBJETIVOS
-----------------------------------------------------------
- Comprender cómo se generan números pseudoaleatorios.
- Implementar un generador congruencial lineal.
- Analizar propiedades de secuencias pseudoaleatorias.
- Generar variables aleatorias con distintas distribuciones.
- Aplicar lo aprendido en una simulación sencilla.

===========================================================
"""

# =========================================================
# IMPORTACIONES
# =========================================================

import random
import math
import matplotlib.pyplot as plt


# =========================================================
# EJERCICIO 1
# PRIMER CONTACTO CON NÚMEROS ALEATORIOS
# =========================================================

print("\n================================================")
print("EJERCICIO 1")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
1. Ejecute este bloque varias veces.
2. Observe los números generados.
3. Responda:
   - ¿Qué rango tienen los números?
   - ¿Se repiten?
   - ¿Parecen aleatorios?
"""

numeros = []

for i in range(10):
    numeros.append(random.random())

print("Números generados:")
print(numeros)


# =========================================================
# EJERCICIO 2
# USO DE SEMILLAS
# =========================================================

print("\n================================================")
print("EJERCICIO 2")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
1. Investigue qué es una semilla (seed).
2. Complete el código.
3. Ejecute varias veces.
4. Observe qué ocurre.

PREGUNTAS:
-----------------------------------------------------------
- ¿Por qué los números ahora son iguales?
- ¿Para qué sirve una semilla?
- ¿Por qué esto es útil en simulación?
"""

random.seed(22)

numeros_seed = []

for i in range(5):
    numeros_seed.append(random.random())

print("Números con semilla:")
print(numeros_seed)


# =========================================================
# EJERCICIO 3
# GENERADOR CONGRUENCIAL LINEAL (GCL)
# =========================================================

print("\n================================================")
print("EJERCICIO 3")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
Complete la implementación del generador congruencial lineal.

FÓRMULA:
X(n+1) = (a * Xn + c) mod m

PREGUNTAS:
-----------------------------------------------------------
- ¿Qué hace el operador %?
- ¿Qué ocurre si cambia la semilla?
- ¿Qué ocurre si cambia el módulo?
- ¿Los números parecen uniformes?
"""


def gcl(seed, a, c, m, n):

    numeros = []

    x = seed

    for i in range(n):

        x = (a * x + c) % m

        u = x / m

        numeros.append(u)

    return numeros


resultado = gcl(
    seed=10,
    a=5,
    c=3,
    m=16,
    n=20
)

print("Secuencia generada:")
print(resultado)


# =========================================================
# EJERCICIO 4
# HISTOGRAMA
# =========================================================

print("\n================================================")
print("EJERCICIO 4")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
1. Genere un histograma usando la secuencia del GCL.
2. Observe la distribución.

PREGUNTAS:
-----------------------------------------------------------
- ¿La distribución parece uniforme?
- ¿Qué significa uniforme?
- ¿Existen huecos o acumulaciones?
"""

plt.figure()
plt.hist(resultado, bins=10, edgecolor='black')
plt.title("Histograma GCL")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()


# =========================================================
# EJERCICIO 5
# GRÁFICO DE DISPERSIÓN
# =========================================================

print("\n================================================")
print("EJERCICIO 5")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
1. Cree un scatter plot.
2. Compare valores consecutivos.

PREGUNTAS:
-----------------------------------------------------------
- ¿Observa patrones?
- ¿Qué significa independencia?
- ¿Por qué patrones visibles son peligrosos?
"""

x = resultado[:-1]
y = resultado[1:]

plt.figure()
plt.scatter(x, y, color='blue', alpha=0.7)
plt.title("Scatter Plot GCL")
plt.xlabel("Xn")
plt.ylabel("Xn+1")
plt.show()


# =========================================================
# EJERCICIO 6
# DISTRIBUCIÓN EXPONENCIAL
# =========================================================

print("\n================================================")
print("EJERCICIO 6")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
Implemente una variable aleatoria exponencial usando:

X = -(1/lambda) * ln(1-U)

donde:
- U es uniforme entre 0 y 1
- lambda > 0

INVESTIGUE:
-----------------------------------------------------------
- ¿Qué representa lambda?
- ¿Dónde se usa esta distribución?
- ¿Por qué aparece el logaritmo?
"""


def exponencial(lamda):

    u = random.random()

    x = -(1/lamda) * math.log(1 - u)

    return x


datos_exponenciales = []

for i in range(1000):
    datos_exponenciales.append(exponencial(0.5))

print("Primeros valores:")
print(datos_exponenciales[:10])


# =========================================================
# EJERCICIO 7
# HISTOGRAMA EXPONENCIAL
# =========================================================

print("\n================================================")
print("EJERCICIO 7")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
1. Grafique un histograma de los datos exponenciales.
2. Compare con la distribución uniforme.

PREGUNTAS:
-----------------------------------------------------------
- ¿La forma cambia respecto al uniforme?
- ¿Por qué muchos valores son pequeños?
- ¿Qué fenómenos reales podrían modelarse?
"""

plt.figure()
plt.hist(datos_exponenciales, bins=30, edgecolor='black')
plt.title("Distribución Exponencial")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()


# =========================================================
# EJERCICIO 8
# COMPARACIÓN ENTRE DISTRIBUCIONES
# =========================================================

print("\n================================================")
print("EJERCICIO 8")
print("================================================")

"""
INSTRUCCIONES:
-----------------------------------------------------------
Calcule:
- media
- mínimo
- máximo

para:
1. distribución uniforme
2. distribución exponencial

PREGUNTAS:
-----------------------------------------------------------
- ¿Qué diferencias encuentra?
- ¿Cuál tiene mayor dispersión?
- ¿Qué aplicaciones tendría cada una?
"""

media_uniforme = sum(resultado) / len(resultado)
min_uniforme = min(resultado)
max_uniforme = max(resultado)

media_exp = sum(datos_exponenciales) / len(datos_exponenciales)
min_exp = min(datos_exponenciales)
max_exp = max(datos_exponenciales)

print("\n--- Estadísticas Distribución Uniforme ---")
print(f"Media: {media_uniforme:.4f}")
print(f"Mínimo: {min_uniforme:.4f}")
print(f"Máximo: {max_uniforme:.4f}")

print("\n--- Estadísticas Distribución Exponencial ---")
print(f"Media: {media_exp:.4f}")
print(f"Mínimo: {min_exp:.4f}")
print(f"Máximo: {max_exp:.4f}")


# =========================================================
# EJERCICIO 9
# MINI SIMULACIÓN
# =========================================================

print("\n================================================")
print("EJERCICIO 9")
print("================================================")

"""
ESCENARIO:
-----------------------------------------------------------
Un pequeño banco quiere simular tiempos de llegada
de clientes.

INSTRUCCIONES:
-----------------------------------------------------------
1. Genere tiempos aleatorios entre 1 y 10 minutos.
2. Simule 100 clientes.
3. Calcule:
   - promedio
   - mínimo
   - máximo

PREGUNTAS:
-----------------------------------------------------------
- ¿Qué representa cada número?
- ¿La simulación parece realista?
- ¿Qué distribución sería mejor?
"""

clientes = 100

tiempos = []

for i in range(clientes):

    t = random.uniform(1, 10)

    tiempos.append(t)

promedio = sum(tiempos) / len(tiempos)
minimo = min(tiempos)
maximo = max(tiempos)

print("\nResultados simulación:")
print(f"Promedio: {promedio:.2f} minutos")
print(f"Mínimo: {minimo:.2f} minutos")
print(f"Máximo: {maximo:.2f} minutos")


# =========================================================
# EJERCICIO 10
# SIMULACIÓN FINAL
# =========================================================

print("\n================================================")
print("EJERCICIO 10")
print("================================================")

"""
TRABAJO FINAL
-----------------------------------------------------------

Diseñe una simulación sencilla usando variables
pseudoaleatorias.

OPCIONES:
-----------------------------------------------------------
- supermercado
- hospital
- tráfico
- videojuegos
- cajero automático
- llamadas telefónicas
- entregas
- otra aprobada por el docente

REQUISITOS:
-----------------------------------------------------------
1. Debe usar generación pseudoaleatoria.
2. Debe usar al menos una distribución.
3. Debe incluir gráficos.
4. Debe incluir interpretación.
5. Debe explicar supuestos y limitaciones.

ENTREGABLE:
-----------------------------------------------------------
- Código funcional
- Explicación del problema
- Resultados
- Conclusiones
"""

print("\n--- SIMULACIÓN: CAJERO AUTOMÁTICO ---")

num_clientes = 100
lambda_llegadas = 0.5
servicio_min = 1
servicio_max = 5

tiempo_actual = 0
tiempos_espera = []
proximo_fin = 0

for i in range(num_clientes):
    llegada = -math.log(1 - random.random()) / lambda_llegadas
    tiempo_actual += llegada

    if tiempo_actual < proximo_fin:
        espera = proximo_fin - tiempo_actual
        tiempos_espera.append(espera)
        inicio = proximo_fin
    else:
        tiempos_espera.append(0)
        inicio = tiempo_actual

    servicio = random.uniform(servicio_min, servicio_max)
    proximo_fin = inicio + servicio

print(f"Clientes: {num_clientes}")
print(f"Tiempo total: {tiempo_actual:.2f} min")
print(f"Espera promedio: {sum(tiempos_espera)/len(tiempos_espera):.2f} min")
print(f"Espera máxima: {max(tiempos_espera):.2f} min")
print(f"Clientes que esperaron: {sum(1 for e in tiempos_espera if e > 0)}")

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(tiempos_espera, bins=20, edgecolor='black')
plt.title("Histograma de Tiempos de Espera")
plt.xlabel("Minutos")
plt.ylabel("Frecuencia")

plt.subplot(1, 2, 2)
plt.plot(range(1, num_clientes + 1), tiempos_espera, 'o-', markersize=3)
plt.title("Evolución de Tiempos de Espera")
plt.xlabel("Cliente")
plt.ylabel("Minutos")
plt.tight_layout()
plt.show()

print("\nSupuestos: llegadas exponenciales, servicio uniforme, cola infinita")
print("Limitaciones: un solo servidor, no hay horas pico")


# =========================================================
# REFLEXIÓN FINAL
# =========================================================

print("\n================================================")
print("REFLEXIÓN FINAL")
print("================================================")

print("\nLaboratorio finalizado.")
```
